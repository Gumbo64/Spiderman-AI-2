```python
import random
from math import pi
import taichi as ti

ti.init(arch=ti.cpu)

n = 8
u = ti.Vector.field(n=2,dtype=ti.f32, shape=(), needs_grad=True)
L = ti.field(dtype=ti.f32, shape=(), needs_grad=True)

@ti.func
def F_spring(vec):  # Force of spring, see https://youtu.be/FcnvwtyxLds?t=612
    spring_k=1
    l=10
    return spring_k*( vec.norm(1e-6) - l) * vec.normalized(1e-6) 



@ti.kernel
def reduce():
    total = u[None]
    v = ti.Vector([0.0,0.0])
    for i in ti.static(range(10)):
        v += F_spring(ti.Vector([120.0,60.0]) - total).norm(1e-6)
        total += v
        
    L[None] = (250 - total.x) ** 2 + (280 - total.y) ** 2


@ti.kernel
def gradient_descent():
    for i in ti.static(range(2)):
        u[None][i] -= u.grad[None][i] * 1


def main():
    # Initialize vectors
    for i in range(2):
        u[None][i] = random.random() * 1000
        

    # Optimize with 100 gradient descent iterations
    for k in range(100000):
        with ti.ad.Tape(loss=L):
            reduce()
#         print('Loss =', L[None])
        gradient_descent()
        print([u[None],(u[None] - ti.Vector([120.0,60.0])).norm(1e-6)])
#         print(u.grad[None])

#     for i in range(n):
#         # Now you should approximately have x[i] == y[i]
#         print(x[i], y[i])


if __name__ == '__main__':
    main()
```
