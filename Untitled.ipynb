```python
import pandas as pd
import numpy as np
from pathlib import Path
import random

from sklearn.neighbors import NearestNeighbors
class MEMORY():
    def __init__(self,mem_size=100000,n_neighbours=10):
        self.mem_size = mem_size
        self.states = np.array([[0 for x in range(52)]])
        self.q = np.array(0)
        
        self.neigh = NearestNeighbors(n_neighbors=n_neighbours)
        self.neigh.fit(self.states)
        
    def add(self,new_state,reward):
        self.states = np.vstack([self.states,[new_state]])
        self.q = np.vstack([self.q,[reward]])
        
        self.states = self.states[-self.mem_size:]
        self.q = self.q[-self.mem_size:]
        
        self.neigh.fit(self.states)
    
    def neighbours(self,new_state):
        return self.neigh.kneighbors(np.array(new_state).reshape(1, -1))

def weights(distances):
    return list(map(lambda d: 1/(1+d**2),distances))

def prob(weights):
    total = np.sum(weights)
    return np.array(list(map(lambda x: x/total , weights)))



def get_V(state,action,neighb):
    mem = memory[action]
    distances, indexes = neighb
    w = weights(distances)
    print(mem.q)
    q = np.array([mem.q[i] for i in indexes])
    p = prob(w)
    return np.sum(np.multiply(q,p)),q



    
    
def get_greedy(state):
    action = 0
    V = -999999
    p = None
    neighb = None
    for i in range(9):
        new_neighb = memory[i].neighbours(observation)
        new_V,new_p = get_V(state,i,neighb)
        if new_V > V:
            action = i
            V = new_V
            p = new_p
            neighb = new_neighb
    return action,V,p,neighb
    
    
    


action_book = {
    0:[0,0,False],
    1:[0,1,True],
    2:[0,-1,True],
    3:[1,1,True],
    4:[1,0,True],
    5:[1,-1,True],
    6:[-1,1,True],
    7:[-1,0,True],
    8:[-1,-1,True],
}



    
# distances, indexes = memory.neighbours(np.array(observation).reshape(1, -1))



```

```python

EPSILON = 0.2
N_ACTIONS = 9
K = 5
GAMMA = 0.99
ALPHA = 0.9

memory = [MEMORY(300000,n_neighbours=K) for i in range(N_ACTIONS)]
```

```python
# fill some memory so that knn has neighbours to use
for action in range(N_ACTIONS):
    observation = np.nan_to_num(env.reset())
    for i in range(K):
        observation, reward, done, info = env.step(action_book[action])
        observation = np.nan_to_num(observation)
        memory[action].add(observation,reward)

        
```

```python
running = True
while running:
    try:
        observation = np.nan_to_num(env.reset())
        neighb = memory[0].neighbours(observation)
        old_neighb=neighb
        done = False
        while not done:
            if random.random() < EPSILON:
                action = random.randint(0,8)
                V,p = get_V(observation,action,neighb)
                neighb = memory[action].neighbours(observation)
            else:
                action,V,p,neighb = get_greedy(observation)
            
            observation, reward, done, info = env.step(action_book[action])
            observation = np.nan_to_num(observation)
            old_neighb = neighb
            neighb = memory[action].neighbours(observation)
            
            
#             learning part
            print("learning part")
            _, max_V_dash ,p , _ = get_greedy(observation,neighb)
            
            delta = reward + GAMMA * max_V_dash - V
            
            distances , indexes = neighb
            
            print(indexes)
            for i in range(len(indexes)):
                ind = indexes[i]
                memory[action].q[ind] += ALPHA * delta * p[i]
            
            memory[action].add(observation,reward)
            
    except KeyboardInterrupt:
        running = False
```
